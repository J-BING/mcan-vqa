```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
data = pd.read_csv('subsetresults.csv')
data.values
```

```python
y1 = [float(item[0].split()[1]) for item in data.values]
y2 = [float(item[0].split()[2]) for item in data.values]
x = [item[0].split()[0] for item in data.values]
```

```python
plt.figure(figsize=(10,5))
plt.title('Results of baseline(glove) and N-KBSN(m) on VQA')
plt.xlabel("Train Subset")
plt.ylabel("Accuracy")
plt.plot_date(x,y1,'-',label="baseline(glove)")
plt.plot_date(x,y2,'-',color='r',label="N-KBSN(m)")
plt.legend()
plt.grid()
```
